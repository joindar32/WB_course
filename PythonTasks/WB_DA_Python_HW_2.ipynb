У нас есть данные о потреблении алкоголя в мире https://github.com/fivethirtyeight/data/tree/master/alcohol-consumption.

Загрузите их и изучите, где же люди пьют больше вина, пива и других спиртных напитков.

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('drinks.csv')
df = data.copy()
```

```python
df.head(10)
```

```python
df.info()
```

```python
index_country_max_beer = df['beer_servings'].idxmax()
country_max_beer = df.iloc[index_country_max_beer]['country']

index_country_max_spirit = df['spirit_servings'].idxmax()
country_max_spirit = df.iloc[index_country_max_spirit]['country']

index_country_max_wine = df['wine_servings'].idxmax()
country_max_wine = df.iloc[index_country_max_wine]['country']

index_country_max_alchohol = df['total_litres_of_pure_alcohol'].idxmax()
country_max_alchohol = df.iloc[index_country_max_alchohol]['country']

print(f'Страна с самым большим количеством потребления пива: {country_max_beer}')
print(f'Страна с самым большим количеством потребления крепких спиртных напитков: {country_max_spirit}')
print(f'Страна с самым большим количеством потребления вина: {country_max_wine}')
print(f'Страна с самым большим количеством потребления чистого алкоголя (спирта): {country_max_alchohol}')
```
